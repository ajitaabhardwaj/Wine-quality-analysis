```python
from sklearn.model_selection import train_test_split, GridSearchCV,cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler,Normalizer, RobustScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
X=df.iloc[:,:-2]  #x are the features used by the model (till 3rd last column)
Y=df.iloc[:,-1:]  #y id the target feature (row vector of last column)
X.head()
```

```python
#scaling of data - robust Scaler removes the median and scales the data according to the quantile range
#standardisation - makes mean 0 and variance 1.
sc=RobustScaler()  
data=sc.fit_transform(X) # The fit method is calculating the mean and variance of each of the features present in 
         #our data. The transform method is transforming all the features using the respective mean and variance.
data[:5]
```

```python
#train-test-split
train_x,test_x,train_y,test_y=train_test_split(data,Y,test_size=0.15,random_state=2)
print("shape of train input data:",train_x.shape,"\n shape of train output data",train_y.shape,
      "\nshape of test input data ",test_x.shape,"\nshape of test output data",test_y.shape)
```

```python
model=LogisticRegression()
model.fit(train_x,train_y)
```

```python
pred=model.predict(test_x)
pred[:5]
```

```python
print("Accuracy Score:",accuracy_score(pred,test_y))
print("classification Report:\n",classification_report(pred,test_y))
print("confusion Matrix:\n",confusion_matrix(pred,test_y))
```

```python
svc=SVC()
svc.fit(train_x,train_y)
```

```python
pred=svc.predict(test_x)
pred[:5]
```

```python
print("Accuracy Score:",accuracy_score(pred,test_y))
print("classification Report:\n",classification_report(pred,test_y))
print("confusion Matrix:\n",confusion_matrix(pred,test_y))
```

```python
svc2=SVC(C= 1, gamma= 0.9, kernel='rbf')  #default C=1.0.
svc2.fit(train_x,train_y)
```

```python
pred=svc2.predict(test_x)
pred[:5]
```

```python
print("Accuracy Score:",accuracy_score(pred,test_y))
print("classification Report:\n",classification_report(pred,test_y))
print("confusion Matrix:\n",confusion_matrix(pred,test_y))
```

```python
#cross validation  
#a resampling procedure used to evaluate machine learning models on a limited data sample. The procedure has a 
#single parameter called k that refers to the number of groups that a given data sample is to be split into.
```

```python
svc_eval = cross_val_score(estimator = svc2, X = train_x, y = train_y, cv =5)
svc_eval.mean()
```

```python
# Consider the inpur varabile as X and Output variable as Y 
X = df[['fixed acidity','citric acid','residual sugar','sulphates','alcohol']]
Y = df[['quality']]
print(X.shape)
print(Y.shape)
print(X)
print(Y)
```

```python
norm=(X-X.min())/(X.max()-X.min())
norm
```

```python
#KNN
classifier = KNeighborsClassifier(n_neighbors=5)  
classifier.fit(norm,Y)
```

```python
#prediction
knnpred= classifier.predict(norm)
knnpred
```

```python
#confusion matrix
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(Y, knnpred))
```

```python
#accuracy
from sklearn.metrics import accuracy_score 
Accuracy_Score = accuracy_score(Y, knnpred)
Accuracy_Score
```
